```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(1)
df = pd.DataFrame({'age':np.random.randint(0,100,size = 15)})
df
```

```python
# define bins
# i'm defining age bins based on 0-12, 12-22, 22-35 etc.
bins = [0,12,22,35,60,70,100]
group_names = ['Kids', 'Young Lads and ladies', 'Young men and women', 'Men and women','Old men and women','Super old']
```

```python
df['age_groups'] = pd.cut(df['age'], bins, labels=group_names)
df
```
